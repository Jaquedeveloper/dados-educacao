# Cadastro de Escolas Municipais, públicas e privadas

Cadastro de escolas municipais, conveniadas e privadas, tanto ativas como extintas, com informações sobre a quantidade de turnos de funcionamento da escola e geolocalização. Dados do Sistema EOL, da Secretaria Municipal de Educação

```python
import pandas as pd
import numpy as np
import pandas_profiling
```

```python
cadastro = pd.read_csv('http://dados.prefeitura.sp.gov.br/dataset/8da55b0e-b385-4b54-9296-d0000014ddd5/resource/a6d8dba5-5959-4e57-903a-09e127e228a5/download/escolasr34.csv',
                       sep=';'
                      )
```

```python
to_date = ['DOM_CRIACAO', 'DT_AUTORIZA', 'DT_CRIACAO', 'DT_INI_FUNC']
for column in to_date:
    cadastro[column] = pd.to_datetime(cadastro[column], errors='coerce')
    
to_text = ['CODESC', 'NOMESC', 'ENDERECO', 'NUMERO', 'BAIRRO', 'CEP', 'TEL1', 'TEL2', 'FAX', 'CODDIST', 'CODINEP',
           'CODCIE', 'EH', 'ATO_CRIACAO', 'NOME_ANT', 'LATITUDE', 'LONGITUDE']

for column in to_text:
    cadastro[column] = cadastro[column].astype(str, errors='ignore')
```
